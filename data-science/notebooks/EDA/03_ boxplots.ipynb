## Boxplots
Nesta etapa, utilizamos boxplots para visualizar a distribuição das variáveis numéricas de forma resumida. Esse tipo de gráfico permite observar a mediana, os quartis, a dispersão dos dados e a presença de valores extremos, complementando as informações já identificadas nas estatísticas descritivas e nos histogramas.

```python
df_amostra = (
    df.groupby("Country", group_keys=False)
      .apply(lambda x: x.sample(frac=0.05, random_state=42))
      .reset_index(drop=True)
)
```

```python
df_amostra = (
    df.groupby("Country", group_keys=False)
      .sample(frac=0.05, random_state=42)
      .reset_index(drop=True)
)
```

```python
df_amostra.shape
```

```python
df["Country"].value_counts(normalize=True)
df_amostra["Country"].value_counts(normalize=True)
```

```python
df_amostra.to_csv("/content/drive/MyDrive/EDA_AquaSense/amostra.csv", index=False)
```

```python
import plotly.express as px

numerical_cols = [
    "Ammonia (mg/l)",
    "Biochemical Oxygen Demand (mg/l)",
    "Dissolved Oxygen (mg/l)",
    "Orthophosphate (mg/l)",
    "pH (ph units)",
    "Temperature (cel)",
    "Nitrogen (mg/l)",
    "Nitrate (mg/l)",
    "CCME_Values"
]

for col in numerical_cols:
    fig = px.box(
        df_amostra,
        y=col,
        title=f'Boxplot de {col}',
        points='outliers'
    )

    fig.update_layout(
        xaxis_title='',
        yaxis_title=col,
        template='plotly_white'
    )

    fig.show()
```

```python
print("Original:")
print(df["CCME_WQI"].value_counts(normalize=True))

print("\nAmostra:")
print(df_amostra["CCME_WQI"].value_counts(normalize=True))
```

```python
print("Original:")
print(df["CCME_Values"].describe())

print("\nAmostra:")
print(df_amostra["CCME_Values"].describe())
```

> Validamos a representatividade da amostra comparando as distribuições da variável-alvo entre o dataset original e a amostra, observando que as proporções e estatísticas descritivas se mantiveram praticamente idênticas.

Os boxplots evidenciam a distribuição das variáveis, destacando mediana, quartis e presença de outliers. Observa-se que variáveis como Ammonia, BOD, Orthophosphate, Nitrogen e Nitrate apresentam forte assimetria à direita, com grande concentração de valores baixos e muitos valores extremos superiores, indicando alta variabilidade.

Já variáveis como Dissolved Oxygen, pH e Temperature apresentam distribuição mais concentrada, com menor dispersão relativa, embora ainda possuam alguns outliers.

A variável CCME_Values se destaca por apresentar maior concentração em valores altos, com menor influência de extremos superiores, refletindo predominância de classificações de melhor qualidade da água.

De modo geral, os outliers observados não devem ser tratados automaticamente como erros, pois podem representar variações reais do ambiente, sendo importante considerar esse comportamento nas etapas de pré-processamento e modelagem.
