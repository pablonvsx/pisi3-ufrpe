# Distribuição das variáveis

## Histogramas
Nesta etapa, utilizamos histogramas para visualizar a distribuição das variáveis numéricas, permitindo identificar padrões de concentração, assimetrias e possíveis caudas longas. Essa análise complementa as estatísticas descritivas ao oferecer uma visão visual do comportamento dos dados.

---

**Objetivo**


*   Visualizar a distribuição das variáveis
*   Identificar padrões de concentração


*   Detectar assimetrias na distribuição
*   Observar possíveis valores extremos









```python
df_amostra = (
    df.groupby("Country", group_keys=False)
      .sample(frac=0.05, random_state=42)
      .reset_index(drop=True)
)
```

```python
df_amostra.shape
```

```python
df["Country"].value_counts(normalize=True)
df_amostra["Country"].value_counts(normalize=True)
```

```python
print("Original:")
print(df["CCME_WQI"].value_counts(normalize=True))

print("\nAmostra:")
print(df_amostra["CCME_WQI"].value_counts(normalize=True))
```

```python
print("Original:")
print(df["CCME_Values"].describe())

print("\nAmostra:")
print(df_amostra["CCME_Values"].describe())
```

```python
numerical_cols = [
    'Ammonia (mg/l)', 'Biochemical Oxygen Demand (mg/l)', 'Dissolved Oxygen (mg/l)',
    'Orthophosphate (mg/l)', 'pH (ph units)', 'Temperature (cel)',
    'Nitrogen (mg/l)', 'Nitrate (mg/l)', 'CCME_Values'
]

for col in numerical_cols:
    fig = px.histogram(df_amostra, x=col, title=f'Distribuição de {col}')
    fig.show()
```

Os histogramas mostram que grande parte das variáveis apresenta distribuição assimétrica à direita, com forte concentração de valores em faixas mais baixas e presença de caudas longas em direção a valores mais elevados. Esse comportamento é evidente em variáveis como Ammonia, Biochemical Oxygen Demand, Orthophosphate, Nitrogen e Nitrate, indicando que a maior parte dos registros está concentrada em níveis menores, enquanto poucos valores extremos ampliam a distribuição.

Já variáveis como Dissolved Oxygen, pH e Temperature apresentam maior concentração em faixas específicas, sugerindo comportamento mais estável, embora ainda existam alguns valores afastados do padrão principal.

No caso de CCME_Values, observa-se forte concentração em valores altos, especialmente próximos de 100, o que sugere predominância de registros associados a melhor qualidade da água.

De forma geral, os histogramas reforçam que a média isoladamente não é suficiente para representar adequadamente o comportamento dos dados, sendo necessário considerar também a distribuição, a assimetria e a presença de valores extremos nas etapas seguintes da análise.
