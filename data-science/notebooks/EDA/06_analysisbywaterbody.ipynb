# Análise por Corpo Hídrico

## Contagem de amostras por tipo de corpo hídrico
Vamos analisar a distribuição dos dados por tipo de corpo hídrico para compreender a representatividade de cada categoria no dataset.

---
**Pergunta que queremos responder**


*   Quantas amostras existem para cada tipo de corpo hídrico?


---
**Objetivo**


*   identificar quais tipos de corpos hídricos são mais ou menos representados no dataset  
*   avaliar como essa distribuição pode influenciar as análises de qualidade da água







```python
if 'Waterbody Type' in df.columns:
    waterbody_counts = df['Waterbody Type'].value_counts()
    print('Contagem de dados por Tipo de Corpo Hídrico:')
    display(waterbody_counts)
```

A contagem por tipo de corpo hídrico revela um forte desbalanceamento na distribuição das amostras do dataset. Os rios (River) são os mais representados, seguidos por efluentes (Effluent) e lagos (Lake). Em contraste, tipos como estuários (Estuarine), baías (Bay), água do mar (Sea Water), canais (Canal), transicionais (Transitional) e costeiros (Coastal) apresentam quantidade bem menor de registros.

Essa distribuição é importante para contextualizar análises futuras, pois conclusões relacionadas aos tipos mais representados tendem a ter maior suporte estatístico. Já para categorias com baixa representatividade, as inferências devem ser feitas com maior cautela, devido à menor quantidade de dados disponíveis.

## Estatísticas descritivas por tipo de corpo hídrico
Agora, vamos gerar estatísticas descritivas para as variáveis numéricas, agrupadas por tipo de corpo hídrico. Isso nos permitirá identificar diferenças nas características da água entre os distintos ambientes.


---
**Pergunta que queremos responder**


*   Existem diferenças aparentes nos parâmetros físico-químicos entre os diferentes tipos de corpos hídricos?



---
**Objetivo**


*   Comparar as medidas descritivas das variáveis numéricas entre os diferentes tipos de corpos hídricos, identificando padrões, variações e possíveis diferenças nos indicadores de qualidade da água








```python
# Verificar se a coluna de tipo de corpo hídrico existe
if 'Waterbody Type' in df.columns:

    # Selecionar automaticamente as colunas numéricas
    numerical_cols = df.select_dtypes(include=['number']).columns.tolist()

    # Remover colunas auxiliares que não fazem sentido na análise descritiva
    cols_to_remove = ['Year']
    numerical_cols = [col for col in numerical_cols if col not in cols_to_remove]

    # Verificar se ainda restaram colunas numéricas válidas
    if numerical_cols:
        waterbody_stats = df.groupby('Waterbody Type')[numerical_cols].agg(['mean', 'median', 'std'])

        print('Estatísticas Descritivas por Tipo de Corpo Hídrico:')
        display(waterbody_stats)
    else:
        print("Nenhuma coluna numérica válida foi encontrada para a análise.")
else:
    print("A coluna 'Waterbody Type' não foi encontrada no DataFrame.")
```

A análise das estatísticas descritivas por tipo de corpo hídrico revela diferenças importantes entre os ambientes aquáticos. Rios (River), por serem o tipo mais representado no dataset, apresentam ampla variabilidade nos parâmetros analisados, refletindo a heterogeneidade das condições ambientais e dos níveis de impacto observados nesse grupo. Lagos (Lake) tendem a apresentar valores médios mais baixos para alguns poluentes quando comparados a categorias mais impactadas, embora também apresentem variabilidade em determinados parâmetros.

Os tipos associados a maior carga de poluição, como Effluent, Sewage e Drainage, destacam-se por apresentar valores médios mais elevados de parâmetros como Ammonia, Biochemical Oxygen Demand, Orthophosphate e Nitrogen, indicando condições mais críticas de qualidade da água. Entre eles, Sewage apresenta os valores mais elevados em vários indicadores, evidenciando um perfil mais fortemente associado à poluição orgânica e química.

Já categorias como Estuarine, Bay, Marine, Sea Water, Transitional e Coastal apresentam comportamentos mais variados, possivelmente influenciados por características próprias desses ambientes. No entanto, para os tipos com menor número de amostras, as estatísticas descritivas devem ser interpretadas com cautela, pois a baixa representatividade pode reduzir a estabilidade das medidas e aumentar a influência de valores extremos.

De forma geral, a análise por tipo de corpo hídrico mostra que a qualidade da água varia conforme o ambiente analisado, sendo essa distinção importante para contextualizar os dados e orientar interpretações mais adequadas nas etapas seguintes.
