```python
numerical_cols = [
    "Ammonia (mg/l)",
    "Biochemical Oxygen Demand (mg/l)",
    "Dissolved Oxygen (mg/l)",
    "Orthophosphate (mg/l)",
    "pH (ph units)",
    "Temperature (cel)",
    "Nitrogen (mg/l)",
    "Nitrate (mg/l)",
    "CCME_Values"
]

skewness_values = df_amostra[numerical_cols].skew().sort_values(ascending=False)

print("Valores de Assimetria (Skewness) das Variáveis Numéricas:")
display(skewness_values)
```

Os coeficientes de assimetria da amostra proporcional confirmam as observações visuais dos histogramas e boxplots. Variáveis como Ammonia, Biochemical Oxygen Demand, Orthophosphate e Nitrate apresentaram forte assimetria positiva, indicando concentração de valores baixos com presença de cauda longa à direita. Temperature, pH e Nitrogen também mostraram assimetria positiva, porém menos intensa. Já Dissolved Oxygen e CCME_Values apresentaram assimetria negativa.

Embora os valores numéricos sejam diferentes do dataset original, o comportamento geral das distribuições foi mantido. Essa diferença ocorre porque o skewness é sensível a valores extremos, que podem variar entre a base completa e a amostra.

A presença de assimetria indica que parte das variáveis não segue normalidade, podendo exigir transformações (como log ou Box-Cox) para melhorar análises e modelos sensíveis à distribuição dos dados.

## Outliers
A análise visual por meio dos boxplots mostra a presença de outliers em várias variáveis numéricas da amostra proporcional. Esses valores aparecem fora dos limites superiores ou inferiores dos gráficos, indicando observações que se distanciam do comportamento central dos dados.

As variáveis que apresentaram maior quantidade de outliers visuais também tendem a ser aquelas com distribuições mais assimétricas, o que reforça a presença de caudas longas e possíveis valores extremos. Esse comportamento pode estar associado a variações naturais do fenômeno analisado ou a registros atípicos que merecem investigação mais cuidadosa.

A identificação desses outliers é importante porque eles podem influenciar medidas estatísticas, como média e desvio padrão, além de impactar modelos sensíveis a valores extremos. Por isso, sua presença deve ser considerada nas próximas etapas da análise, avaliando se esses valores representam eventos reais do conjunto de dados ou se indicam a necessidade de tratamento.
