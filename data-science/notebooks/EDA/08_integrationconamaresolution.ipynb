# Conexão com os parâmetros CONAMA e Próximos Passos

## O que é o CONAMA?
O Conselho Nacional do Meio Ambiente (CONAMA) é o órgão responsável por estabelecer normas e padrões para a qualidade ambiental no Brasil. No contexto de recursos hídricos, destaca-se a Resolução CONAMA nº 357/2005, que define critérios para a classificação e o enquadramento dos corpos de água, bem como os limites aceitáveis para diversos parâmetros físico-químicos e biológicos.

Esses parâmetros são utilizados como referência para avaliar a qualidade da água e determinar se ela atende aos usos previstos, como abastecimento, recreação ou preservação ambiental.



---
Embora o dataset analisado contenha diversas variáveis relacionadas à qualidade da água, ele não possui uma classificação diretamente baseada nos padrões brasileiros.


Dessa forma, o CONAMA será utilizado como:

- Referência normativa para avaliação da qualidade da água
- Base para construção de um rótulo classificatório
- Elemento de conexão entre os dados e a realidade brasileira
- Fundamento para a aplicação de técnicas de Machine Learning


---
Relação entre o dataset e os parâmetros do CONAMA

A análise exploratória realizada permitiu identificar variáveis do dataset que possuem correspondência com parâmetros considerados relevantes pelo CONAMA, tais como:

- Nitrogênio (Nitrogen (mg/l))
- Nitrato (Nitrate (mg/l))
- Amônia (Ammonia (mg/l))
- Demanda Bioquímica de Oxigênio (Biochemical Oxygen Demand (mg/l))
- pH (pH (ph units))

Além disso, a análise de correlação indicou que variáveis relacionadas à carga de nutrientes, como ortofosfato e nitrogênio, apresentam associação com o índice de qualidade (CCME_Values), reforçando sua relevância para a avaliação da qualidade da água.

Os resultados da análise exploratória indicaram que variáveis associadas à presença de nutrientes e poluição apresentam maior relação com a qualidade da água no dataset.

Esse comportamento está alinhado com os parâmetros utilizados pelo CONAMA, que também considera esses compostos como indicadores importantes para a classificação dos corpos hídricos.

Assim, a EDA não apenas permitiu compreender os dados, mas também ajudou a identificar quais variáveis são mais relevantes para a construção de um modelo baseado em critérios normativos.

Com base nas análises realizadas, os próximos passos incluem:

- Definir a lógica de classificação baseada nos parâmetros do CONAMA
- Construir uma nova variável alvo (rótulo) com base nesses critérios
- Adaptar os dados do dataset para refletir essa classificação
- Aplicar modelos de Machine Learning para prever a qualidade da água
- Integrar os resultados ao contexto do aplicativo AquaSense

A análise exploratória realizada teve como objetivo compreender a estrutura, a qualidade e o comportamento das variáveis presentes no dataset, servindo como base para decisões futuras relacionadas à modelagem e à aplicação no contexto do projeto.

Inicialmente, foi realizada uma visão geral do dataset, permitindo identificar suas principais características, como o tipo de dados disponíveis, a diversidade de variáveis e o contexto das informações relacionadas à qualidade da água. Em seguida, a análise de qualidade dos dados indicou a ausência de valores nulos, garantindo consistência estrutural e permitindo avançar para etapas mais analíticas sem necessidade de tratamentos iniciais complexos.

A etapa de estatísticas descritivas possibilitou compreender medidas centrais e de dispersão das variáveis, enquanto a análise das distribuições revelou padrões importantes, como assimetrias e concentrações de valores, que indicam comportamentos não uniformes entre os parâmetros avaliados. Essas observações foram fundamentais para entender a natureza dos dados e possíveis implicações em análises futuras.

Na sequência, foram realizadas análises segmentadas por país e por corpo hídrico, permitindo identificar diferenças na distribuição dos dados e possíveis variações associadas ao contexto geográfico. Essas análises evidenciaram que o dataset não é homogêneo, apresentando variações relevantes entre diferentes regiões e agrupamentos, o que deve ser considerado em etapas posteriores.

A análise de correlação entre variáveis revelou que a maioria das relações lineares é fraca, porém destacou associações relevantes entre variáveis relacionadas à presença de nutrientes, como ortofosfato, nitrogênio e nitrato. Essas variáveis apresentaram correlação negativa com o índice CCME_Values, sugerindo que o aumento desses compostos está associado à redução da qualidade da água. A seleção dos pares mais relevantes permitiu focar nas relações mais expressivas, e os gráficos de dispersão mostraram que, em alguns casos, essas relações não são estritamente lineares, evidenciando comportamentos mais complexos do que os indicados apenas pelos coeficientes de correlação.

A EDA indicou que variáveis associadas à carga de nutrientes e à poluição apresentam maior relevância na variação da qualidade da água dentro do dataset, enquanto outras variáveis, como pH, temperatura e oxigênio dissolvido, demonstraram menor associação linear.

Com base nesses resultados, foi possível estabelecer uma conexão com os parâmetros definidos pelo CONAMA, especialmente no que diz respeito à avaliação da qualidade da água por meio de variáveis físico-químicas. A identificação de correspondência entre variáveis do dataset e parâmetros normativos reforça a viabilidade de utilizar o CONAMA como referência para a construção de uma classificação adaptada à realidade brasileira.

Em conclusão, a análise exploratória não apenas permitiu compreender os dados, mas também forneceu direcionamento para as próximas etapas do projeto. Como continuidade, pretende-se estruturar uma lógica de classificação baseada nos critérios do CONAMA, construir uma variável alvo representando a qualidade da água e aplicar técnicas de Machine Learning para modelar e prever essa classificação, integrando os resultados ao contexto do aplicativo proposto.
