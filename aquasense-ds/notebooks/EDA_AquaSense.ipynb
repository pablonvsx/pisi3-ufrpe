## Contexto
Este notebook apresenta a Análise Exploratória de Dados (EDA) de um dataset de qualidade da água, utilizado no contexto do projeto AquaSense.

O objetivo desta análise é compreender o comportamente dos dados antes de qualquer modelagem, buscando identificar padões, incosistências e limitações do dataset.

- Entender a estrutura e composição dos dados
- Identificart problemas de qualidade (valores nulos, inconsistências, etc.)
- Analisar distribuições e possíveis outliers
- Explorar relações entre variáveis
- Identificar parâmetros compatíveis com a Resolução CONAMA 357/2005

Essa estapa é fundamental para orientar decisões futuras, especialmente na construção de um rótulo classificatório adaptado ao contexto brasileiro.
